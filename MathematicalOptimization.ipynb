### Mathematical Optimization

It all starts with sets and functions. If sets are a collection of distinct objects, functions are the maps from one set (inputs) to another set (outputs). In an optimization problem, we're trying to pick the best possible solution from the set of all feasible solutions, according to some objective function.

The gradient method is perhaps the most simple and well known algorithm for solving an optimization problem. Consider the following unconstrained optimization problem: $$\min f(x)$$ where the objective function is differentiable. $$\nabla f$$


Consider the following optimization problem:$$\min f(x, y) = 10x^2 + 15y^2 + e^{x+ y}$$

$$\nabla f= \begin{pmatrix}
20x + e^{x + y}\\
30y + e^{x + y}
\end{pmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x: float, y: float) -> float:
    return 10 * x ** 2 + 15 * y ** 2 + np.exp(x + y)

def dx(x: float, y: float) -> float:
    return 20 * x + np.exp(x + y)

def dy(x: float, y: float) -> float:
    return 30 * y + np.exp(x + y)
```

```python
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax = plt.axes(projection='3d')
ax.plot_surface(
    X, Y, Z, rstride=1, cstride=1,
    cmap='viridis', edgecolor='none'
)
plt.show()
```

```python
x = 10
y = 10
t = 0.03

for k in range(50):
    # get gradient values
    _dx = dx(x, y)
    _dy = dy(x, y)

    x += -t * _dx
    y += -t * _dy

    print(f'X: {x}, Y: {y}')
```
